```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500.0, 480.0, 600.0, 750.0, 900.0, 850.0, 920.0, 1100.0, 1300.0, 1350.0, 1500.0, 1450.0, 1550.0, 1600.0, 1650.0, 1600.0, 1500.0, 1400.0, 1600.0, 1700.0, 1800.0, 1750.0, 1850.0, 1900.0],
    'Product B Sales': [300.0, 320.0, 400.0, 450.0, 500.0, 580.0, 700.0, 750.0, 820.0, 900.0, 950.0, 980.0, 1050.0, 1100.0, 1150.0, 1200.0, 1250.0, 1300.0, 1350.0, 1400.0, 1500.0, 1600.0, 1550.0, 1700.0],
    'Product C Sales': [200.0, 210.0, 250.0, 280.0, 320.0, 350.0, 380.0, 400.0, 420.0, 440.0, 460.0, 480.0, 500.0, 520.0, 540.0, 560.0, 580.0, 600.0, 620.0, 640.0, 660.0, 680.0, 700.0, 720.0]
}

sales_df = pd.DataFrame(data)

totalsales = sales_df.sum()
print(totalsales)

averagesales = sales_df.mean()
print(averagesales)

for col in sales_df.columns[1:]:
    max_month = sales_df[col].idxmax()
    max_sales = sales_df[col].max()
    print(f"The month with the highest sales for {col} is {max_month} with {max_sales} units sold.")




sales_df['Month'] = pd.to_datetime(sales_df['Month'])
sales_df.set_index('Month', inplace=True)

plt.plot(sales_df.index, sales_df['Product A Sales'], label='Product A')
plt.plot(sales_df.index, sales_df['Product B Sales'], label='Product B')
plt.plot(sales_df.index, sales_df['Product C Sales'], label='Product C')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for Each Product (2021-2022)')

plt.legend()
plt.show()

corr = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])
print("\n",f"The correlation coefficient between Product A and Product B sales is {corr:.2f}.")
```
